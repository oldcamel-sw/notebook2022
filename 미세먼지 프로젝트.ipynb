# 1. 기상 데이터를 통한 미세먼지 예측

미세먼지를 관측하는 단위인 PM10은 "직경 10마이크로미터 이하의 먼지"를 뜻한다.
- 미세먼지는 단일 물질이 아니라 다양한 생성과정에 의해 만들어지는 미세한 고체와 액체 입자로, 대기 중에 떠 있는 아주 작은 먼지다.
- 생성원에 따라 1차/2차 미세먼지로 나뉜다.
    - 1차: 연료연소시설의 굴뚝, 자동차 배기구 등
    - 2차: 대기 중에서 황산화물, 질산화물, 암모니아가 화학 반응을 통하여 생성하는 유기물질<br><br>
    
    
**미세먼지를 잘 예측하려면 미세먼지 관측 데이터가 많아야 하고, 미세먼지 관측 데이터가 많으려면 미세먼지 측정소가 많아야 한다.**
- 그러나.. 실상은 미세먼지 측정소는 특정 지역에 집중되어 있고; 측정소가 희소한 지역에서는 미세먼지 농도 예측이 어렵다 => 이로 인해 지역 피해가 발생한다 (임준묵, 2019).
- 이에 반해 기상청의 기상관측장비는 전국 곳곳에 위치해 있고, 오랜 기간 동안 실시간으로 쌓인 데이터가 많다.
    - **만약 기상정보로 미세먼지를 예측할 수 있다면 관측재원(측정소)이 없는 곳에서도 미세먼지 농도 예측 / 피해 예방이 가능할 것**이다.

# 2. 가설 목록

**미세먼지에 대해 떠도는 속설 + 전문가들이 밝힌 연구내용들은 다음과 같은 것들이 있다.**
- 비가 많이 오면 먼지가 씻겨 내려가서 미세먼지 농도가 낮아진다. => 강수가 많은 여름철(7-9월)이 강수가 적은 겨울철/봄철(12-5월)보다 미세먼지 농도가 높을 것이다.<br><br>

- 중국의 대기오염 때문에 중국에서 바람이 불어오면 국내 미세먼지 농도가 높아진다 => 즉, 풍향이 서쪽일 때가 동쪽일 때보다 미세먼지 농도가 높다.<br><br>

- 풍속이 빠르면 먼지들이 빠르게 쓸려가서 미세먼지 농도가 낮아진다. => 풍속이 낮을 경우 대기정체로 인해 미세먼지 농도가 높아질 것이다.<br><br>

- 2차 미세먼지 생성원인 황산화물, 질산화물 등의 대기 중 농도가 높을수록 미세먼지 농도도 높을 것이다 => NO2와 SO2가 높을수록 미세먼지 농도가 높을 것이다.<br><br>

> RAIN, SNOW, 강수계속시간, 안개계속시간은 Null값 = 0

- CO는 PM10과 높은 상관관계(0.73)를 지닌 중요한 변수인데, 결측치가 꽤 많았다 (57개).
- 또 눈여겨 본 점은 CO가 NO2(0.85), SO2(0.63), O3(-0.43), ATM_PRESS(0.41)과의 상관관계가 꽤 높다는 것이었다.
    - CO 결측치인 행을 제외한 데이터로 "NO2, O3, SO2, ATM_PRESS"로 CO에 대한 선형회귀모형을 만들었는데, R^2가 0.97이었고; cross validation해본 결과 MSE도 소수점 셋째자리 수준이었다. => CO 값이 소수점 첫째자리이기 때문에, 소수점 셋째자리의 MSE는 거의 오차가 없는 수준이라고 봐도 된다고 판단했다.
    - 이 선형회귀모형을 이용해서 CO 결측치를 채웠다.

```python
# CO 결측치를 채우기 위한 선형회귀모형
df_dropped = df_edited.dropna(axis=0)
lm = sm.OLS(df_dropped['CO'], df_dropped[['NO2', 'O3', 'SO2', 'ATM_PRESS']])
lm = lm.fit()
print('"NO2, O3, SO2, ATM_PRESS"로 CO에 대한 선형회귀 R^2:', lm.rsquared.round(3))
```

```python
# CO 결측치 채우기
CO_null_idx = df_edited[df_edited.isnull()['CO']].index
for idx in CO_null_idx:
    df_edited['CO'].iloc[idx] = lm.predict(df_edited[['NO2', 'O3', 'SO2', 'ATM_PRESS']])[idx].round(1)
```

## 4-2. 변수 라벨링

- 시계열 EDA를 위한 YearMonth 라벨링

```python
# YearMonth 변수 만들기
df_edited['YearMonth'] = df_edited['MeasDate'].apply(lambda x: str(x).split('-')[0]+str(x).split('-')[1])
```

- 16방위 풍향을 4범주로 축소시키면, 풍향이 미세먼지에 미치는 영향이 더 명확히 드러나지 않을까?

```python
# 풍향 변수 라벨링
def get_wind_cat(val):
    if val <= 90:
        wind_cat = 'NE'
    elif (val >= 91) & (val <= 180):
        wind_cat = 'SE'
    elif (val >= 181) & (val <= 270):
        wind_cat = 'SW'
    elif (val >= 271) & (val <= 360):
        wind_cat = 'NW'
        
    return wind_cat

df_edited['WIND_DIR'] = df_edited['WIND_DIR'].apply(get_wind_cat)
```

- 국내 미세먼지 기준 라벨링

```python
# 미세먼지 '좋음, 보통, 나쁨' 범주 만들기
    # 원래 151 이상이 매우나쁨 이지만 데이터셋에 PM10>151 인 관측치가 3개밖에 없어서 '매우나쁨' 범주 사용 안함
def get_PM10_cat(val):
    if val >= 81:
        cat = 'bad' # 나쁨
    elif val >= 31:
        cat = 'normal' # 보통
    else:
        cat = 'good' # 좋음
    return cat

df_edited['PM_CAT'] = df_edited['PM10'].apply(get_PM10_cat)
```

- 미세먼지와 계절의 상관관계 파악을 위한 계절 라벨링

```python
# 계절 변수 만들기
def get_season_cat(val):
    if val >= 12:
        season = 'winter'
    elif val >= 9:
        season = 'autumn'
    elif val >= 6:
        season = 'summer'
    elif val >= 3:
        season = 'spring'
    else:
        season = 'winter'
    return season

df_edited['SEASON'] = df_edited['MeasDate'].apply(lambda x: get_season_cat(int(str(x).split('-')[1])))
```

### ANOVA

```python
spring_pm10 = df_edited[df_edited['SEASON']=='spring']['PM10']
summer_pm10 = df_edited[df_edited['SEASON']=='summer']['PM10']
autumn_pm10 = df_edited[df_edited['SEASON']=='autumn']['PM10']
winter_pm10 = df_edited[df_edited['SEASON']=='winter']['PM10']
```

```python
fstat, pval = stats.f_oneway(spring_pm10, summer_pm10, autumn_pm10, winter_pm10)
print("F-통계량: {0:.3f},\t p-value: {1:.3f}".format(fstat, pval))
```

- **p-value가 0이므로 네 집단 간의 PM10 평균 사이에 통계적으로 유의미한 차이가 있다.**

### Tukey's HSD 사후검정
- 네 집단 사이에 차이가 있는 건 알겠는데, 어느 집단과 어느 집단 사이에 차이가 있는지 알아내기 위한 기법이 사후검정이다.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# np.concatenate로 데이터 합치기
v = np.concatenate([spring_pm10, summer_pm10, autumn_pm10, winter_pm10])

# 데이터 개수만큼 레이블 준비
labels = ['spring'] * len(spring_pm10) + ['summer'] * len(summer_pm10) + ['autumn'] * len(autumn_pm10) + ['winter'] * len(winter_pm10)

# 사후검정 수행
tukey_results = pairwise_tukeyhsd(v, labels, 0.05)
print(tukey_results)
```

> **REJECT가 모두 True이므로, 서로 평균이 같은 그룹이 전혀 없는 것이다. 계절을 기준으로 집단을 묶으면, PM10이 모두 유의미하게 달라진다.**
> - *분명히 계절이 PM10에 영향을 주는 것 같다.*

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='SEASON', y='PM10', data=df_edited)
plt.show()
```

- 서로 비슷해보이지만, Tukey's test 결과에 따라 이 네 집단 사이에 서로 모두 통계적으로 유의미한 차이가 있다고 주장할 수 있다.
    - **겨울이 제일 미세먼지 농도가 높고, 여름이 제일 낮다.**

> - 기존에 세웠던 가설들과 관련된 변수들(계절, 강수량, 풍속, 풍향 등)이 모델링 과정에서도 유의한 변수들로 도출되고 모델의 설명력에 큰 중요성을 차지했더라면, 기존에 세웠던 가설들을 검정하는 것이 더 쉬웠을 것이다.
> - 그러나 EDA/사전학습으로 추측했던 유의변수와 모델링을 통해 도출된 유의변수 간에 차이가 보이면서, 위의 데이터를 직관적으로 이해하기가 더 어려워졌다.
