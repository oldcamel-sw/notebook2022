## 1. Ndarray & Python List

```python
a = np.array([1,2,3])
type(a)
a
```

```python
a = [1,2,3]
b = [4,5,6]
new_list = []
for e1, e2 in zip(a,b):
    new_list.append(e1 + e2)

print(new_list)
print()
print(a + b)
```

## 4. DataFrame
- 다수의 Series를 하나의 변수로 관리할 수 있도록 만든 자료형
- Series의 dict 형태로 보면 됌.
  - {'칼럼1': Series1, '칼럼2': Series2}
  - 각 Series는 DataFrame의 column을 이룸
  - DataFrame을 이루는 Series 간의 index는 서로 다 같음.

### 4.1. How to Create DataFrame

```python
s1 = np.arange(1, 6, 1)
s2 = np.arange(6, 11, 1)
df = pd.DataFrame(
    {
        'c1': s1,
        'c2': s2
    }
)

df = pd.DataFrame(
    [
        [10, 11],
        [10, 12]
    ]
)

df = pd.DataFrame(
    np.array(
        [
            [10, 11],
            [20, 21]
        ]
    )
)

df = pd.DataFrame(
    [
        pd.Series(np.arange(10, 15)),
        pd.Series(np.arange(15, 20))
    ]
)

df = pd.DataFrame(
    [
        np.arange(10, 15),
        np.arange(15, 20)
    ]
)

df = pd.DataFrame(
    np.array(
        [
            [10, 11],
            [20, 21]
        ]
    ),
    columns = ['a', 'b'],
    index = ['r1', 'r2']
)

# 1줄 짜리를 만들더라도 value에 해당하는 값은 iterable dtype 이어야 함
df = pd.DataFrame({'c1': [0], 'c2': [1]})

s1 = pd.Series(np.arange(1,6,1), index=['a','b','c','d','e'])
s2 = pd.Series(np.arange(6,11,1), index=['b','c','d','a','e'])
df = pd.DataFrame(
    {
        'c1': s1,
        'c2': s2
    }
)
```

### 4.2. Reindexing

```python
# set_index() : 특정 column을 index로 만듬
df.set_index('c1')
```

```python
# 진짜 reindex
s = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
s2 = s.reindex(
    ['a', 'c', 'e', 'g']
)
s2
```

## 7. Arithmetic
**연산 기준**
- DataFrame은 Column 기준
- Series는 index 기준
- 따로 명시가 없다면 Series의 index가 DataFrame의 columns에 맞춰짐

```python
import FinanceDataReader as fdr
price_df = fdr.DataReader('005930','2009-09-16','2018-03-21')
price_df.head(3)
```

### 7.1. DataFrame & Series

```python
price_df.iloc[0]
```

```python
# subtract row Series
# DataFrame의 기준인 columns와 Series의 기준인 index가 서로 일치함.
## → 의도한대로 계산됨!
(price_df - price_df.iloc[0]).head(3)
```

```python
# subtract column Series
price_df['Open'].head(3)
(price_df - price_df['Open']).head(3)
```

### 7.2. DataFrame & DataFrame
- index, column이 일치하는 것들 끼리만 element-wise 연산이 이루어지고 나머지는 nan 처리

```python
price_df[['Open','Low']].iloc[2]
```

```python
(price_df - price_df[['Open', 'Low']].iloc[2]).head()
```

### 7.3. Axis

```python
df[['순이익률(%)', 'PER(배)']].sum()  # default axis=0
```

```python
df[['순이익률(%)', 'PER(배)']].sum(axis=1)
```

```python
df[['순이익률(%)', 'PER(배)']].mean()  # default axis=0
```

```python
# DataFrame - Series 형태
# "mean 값을 통해 Normalizing 형태"
(price_df - price_df.mean()).head(3)
```

### 7.4. `.sub()`

```python
# .sub()을 이용하면 가능!
price_df.sub(close_series, axis=0).head(3)
```
