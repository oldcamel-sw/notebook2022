```python
import random
r = random.randint(0,100)   # 0에서 100까지 (100 포함)

while True:
  try:
    guess = int(input("Guess a Number! : "))  # input은 str으로 입력 받음
    
    if guess < r :
      print("Too low!")
    elif guess > r :
      print("Too High!")
    elif guess == r :
      print(f"Just right. The answer is {r}.")
      break
  
  except:
    print("Must type a number!")
```

```python
def mysum(*numbers):
  sum_num = 0

  for num in numbers:
    sum_num += num
  return sum_num

print(mysum(1,4,5,3,9))
print(mysum(*[3,1,6],9,2,4))
```

```python
def pig_latin(word):
  if word[0] in 'aeiou':
    return f'{word}way'
  
  return f'{word[1:]}{word[0]}ay'

print(pig_latin('python'))
```

```python
def pig_latin2(word):
  if word[0].isupper() & word[1:].islower():
    if word[0] in 'aeiou':
      return f'{word}way'
    
    return f'{word[1].upper()}{word[2:]}{word[0].lower()}ay'
  
  else:
    return pig_latin(word)

print(pig_latin2('Python'))
```

```python
def pl_latin(sent):
  new_sent = []
  for s in sent.split():
    new_sent.append(pig_latin(s))
  
  return ' '.join(new_sent)
```
