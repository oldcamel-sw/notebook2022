```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats


plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
mu = 0
std = 1
rv = stats.norm(mu, std)
```

```python
xx = np.linspace(-5, 5, 100)
plt.plot(xx, rv.pdf(xx))
plt.arrow(0, 0.05, 2, 0, lw=3, color='r', 
          head_width=0.02, head_length=0.2, length_includes_head=True)
plt.arrow(0, 0.05, -2, 0, lw=3, color='r', 
          head_width=0.02, head_length=0.2, length_includes_head=True)
plt.text(-1.3, 0.02, "표준편차의 약 4배")
plt.ylabel("p(x)")
plt.title("정규분포의 확률밀도함수")
plt.show()
```

## 왜도 skewness & 첨도 kurtosis
> 왜도 0, 첨도 1일 경우가 완전한 정규분포다.
- 왜도가 양수이면 right skewed (positive skewed)
- 왜도가 음수이면 left skewed (negative skewed)이다.
- 첨도가 양수이면 그래프가 뾰족해지고, 음수가 되면 그래프가 완만해진다.
