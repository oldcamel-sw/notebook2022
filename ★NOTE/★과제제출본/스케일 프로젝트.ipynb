# 1. 후판 공정과 스케일 불량

## 1-1. 주요 내용 / 변수 설명

#### 후판의 정의
후판은 두꺼운 철판을 의미한다.6mm 이상의 열연 강판. 교량, 차량, 구조물, 압력용기, 선박제조 등에 쓰인다.


#### 후판 공정에서 가열과 압연
제강 공정과 연주 공정을 지나 만들어지는 중간재인 slab는 압연 공정까지 이동 도중 온도가 약 740\~800도까지 떨어지게 된다. 압연의 적정 온도는 일반적으로 1000\~1100도 사이이므로, **적정 온도를 맞춰주기 위해 '가열 공정'이 필요**하다.


#### 가열로 구성
가열로의 종류는 예열대, 가열대, 균열대가 있다. 가열대에서 온도를 높여주고, 균열대에서는 slab 내부까지 열이 고르게 퍼지도록 해준다.


#### 스케일 불량이란
스케일은 가열 처리나 압연 처리 중 slab나 billet의 표면에 붙어 있는 산화철(scale)을 의미한다. 가열 중에 생기는 스케일을 1차 스케일, 압연 중에 생기는 스케일을 2차 스케일이라고 한다.
- 스케일은 높은 온도에서 극성을 부린다. 1600도가 넘는 온도로 출강 생산되는 반제품들은 공기와 접촉하는 순간부터 표면에 스케일이 끼기 시작한다.
- 강종마다 스케일 발생 정도가 다른데, 규소강의 경우 스케일 발생량이 적다.

#### 스케일 제거 기법
스케일이 있는 상태에서 압연 및 도금 등 후공정을 하면 제품의 완성도가 떨어진다. 즉, 제품 표면의 불순물이 낀 채로 압연을 하면 표면 크랙이 발생한다. 그래서인지 공정 중에 시행하는 여러 번에 걸쳐서 디스케일링 작업을 시행한다.
- HSB(Hydraulic Scale Breaker): 가열 처리 중 발생하는 스케일을 제거하기 위해, 가열로에서 나온 반제품 표면에 고압수를 뿌려 스케일을 떼어내는 작업이다.
- 조압연기(roughing mill), 사상압연기(finishing mill) 등을 거치는 압연 공정 중에도 수시로 디스케일링 과정을 거친다.

# 2. 데이터 전처리

## 2-1. Feature Engineering

```python
# PLATE_NO는 id값이니 제외
df_edited = df_raw.drop('PLATE_NO', axis=1)

# SCALE 정수 labelling
df_edited['SCALE'] = df_edited['SCALE'].map({'양품': 0, '불량': 1})

# HSB 정수 labelling
df_edited['HSB'] = df_edited['HSB'].map({'미적용': 0, '적용': 1})

# STEEL_KIND labelling
df_edited['STEEL_KIND'] = df_edited['STEEL_KIND'].str[:1]

# SPEC labelling
df_edited['SPEC'] = df_edited['SPEC'].apply(lambda x: x.split('-')[0])
df_edited['SPEC'] = df_edited['SPEC'].apply(lambda x: x.split('/')[0])

# 관측치가 20개 미만인 규격(SPEC)들 => "UNCOMMON"
uncomm_specs = ['A283', 'V42JBN3', 'A516', 'API', 'A709', 'A131', 'CCS']
df_edited['SPEC'] = df_edited['SPEC'].apply(lambda x: 'UNCOMMON' if x in uncomm_specs else x)

# Rolling_Date to datetime
df_edited['ROLLING_DATE'] = df_edited['ROLLING_DATE'].apply(lambda x: ':'.join(x.split(':')[0:2]))
df_edited['ROLLING_DATE'] = pd.to_datetime(df_edited['ROLLING_DATE'], format='%Y-%m-%d:%H')
```

```python
dt_feat = dt_feat.sort_values(by='DT imp', ascending=False)[:10].reset_index(drop=True)

dt_feat
```

```python
lr_feat = pd.DataFrame()
lr_feat['LR feat'] = X_lr_train.columns
lr_feat['LR exp(coef)'] = np.exp(lr.coef_).ravel()
lr_feat1 = lr_feat.sort_values(by='LR exp(coef)', ascending=False).round(3)[:5]
lr_feat2 = lr_feat.sort_values(by='LR exp(coef)', ascending=False).round(3)[-5:]

lr_feat = pd.concat([lr_feat1, lr_feat2], axis=0)
lr_feat = lr_feat.reset_index(drop=True)
lr_feat
```

- 로지스틱 회귀 계수에 exp를 씌워주면 odds ratio가 된다. odds ratio=1 이면 '영향이 없다'이고, 예를 들어 오즈비가 1.5이면 "해당 독립변수가 1단위 증가할 때마다 종속변수가 1이 될 확률이 50% 더 높다"고 해석한다.<br><br>

- 회귀계수가 가장 높은 변수는 압연 온도인데, 압연 온도가 섭씨 1도 증가할 때마다 불량일 확률이 약 3% 증가하는 것이다.
- 그런데 이게 꽤 큰 숫자인 게, 압연온도는 섭씨 1도씩 왔다갔다 하는 변수가 아니라, 최솟값 222부터 최댓값 1078까지, 평균 938에 표준편차는 78이다.
- 예를 들어 압연온도가 표준편차인 78만큼 증가하면, 무려 불량률이 약 2.34배나 증가하는 것이다.

```python
df_selected['ROLLING_TEMP_T5'].describe()
```

- 가장 회귀계수가 낮은 변수는 균열대의 온도/시간 변수인데, 1단위 증가할 때마다 불량률일 확률이 약 1% 미만씩 낮아진다.
- EDA에서 살펴봤듯, 균열대에 반제품이 오래 머물수록 반제품의 온도는 낮아지는 경향이 있다.
    - 이는 압연온도가 낮을 때 불량률이 낮아진다는 결과와 일관성 있는 결과다.
