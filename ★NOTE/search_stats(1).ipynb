## 2장 데이터와 표본분포

**표준오차(Standard Error)** 는 표본의 표준편차(표본표준편차)를 표본수의 제곱근으로 나눈 것이다. (여러 표본들로부터 얻은 "표본통계량"의 변량이다.)
$$ SE = \dfrac{s}{\sqrt n} $$
- 물론 모집단의 표준편차를 아는 경우 $\sigma$를 써도 되지만, 대부분 모르는 경우니까 표본의 표준편차를 넣는다.
- **표준오차는 통계에 대한 표본분포의 변동성을 한마디로 말해주는 단일 측정 지표다**. 
- 표준오차와 표본크기의 관계는 'n제곱근의 법칙' => 표준오차를 2배로 줄이려면 표본크기를 4배로 증가시켜야 한다.
- 일반적으로.. 표준오차를 추정하기 위해 새 샘플을 수집하는 접근방식이 아니라 "부트스트랩 재표본"을 사용한다.<br><br>

**표본오차(Sampling Error)** 는 모집단을 대표할 수 있는 전형적인 구성 요소를 선택하지 못함으로써 발생하는 오차다.
    - "모집단을 대표한다고 할 수 있는 대표성 있는 표본을 뽑는 걸 실패하는 데서 발생하는 오차"를 뜻한다.<br><br>
    - 표본편향(sample bias: 모집단을 잘못 대표하는 표본)이라는 개념과도 관련 있어 보인다.
    
**오차한계=표본오차** 란
- 추정을 할 때 '모평균 추정구간의 중심으로부터 최대한 허용할 최대허용오차'를 말한다.
- 오차한계를 구하는 법은 "임계값(critical value) X 표준오차(SE)"다.
    - 임계값은 표준정규분포일 때는 z값, t분포일 때는 t값, 카이제곱 분포에서는 카이제곱 값이다.
    - e.g. Z값을 사용하는 경우 $ Z \times \dfrac{\sigma}{\sqrt n} $

**표본분포(sampling distribution)** 란 "하나의 동일한 모집단에서 얻은 여러 샘플에 대한 표본통계량의 분포"다.
- 표본을 통한 추정에는 오류가 있다. 왜냐? 표본은 뽑을 때마다 달라지기 때문이다. "우리가 관심 있는 것은 **표본의 변동성(sampling variability)** 이다.
- 평균과 같은 표본통계량의 분포는 데이터 자체의 분포보다 규칙적이고 종 모양일 가능성이 높다. => 표본이 클수록 그럴 가능성이 높다. (중심극한정리)
    - 예를 들어 1000개 값으로 이뤄진 표본; 5개 값의 평균 1000개 값으로 이뤄진 표본, 20개 값의 평균 1000개 값으로 이뤄진 표본을 비교해보면 => n(표본크기)이 커질 수록 표본통계량(여기선 평균)의 분포가 정규분포 형태를 띤다.<br>
> 중심극한정리 덕분에, 추론을 위한 표본분포에, 즉 신뢰구간이나 가설검정을 계산하는 데에 t 분포 같은 정규근사 공식을 사용할 수 있다.

도수분포표, 히스토그램, 상자그림, 표준오차는 모두 표본추정에서 잠재적인 오차를 이해하는 방법들이다. 신뢰구간은 이들과 다른 방법이다.
- 신뢰수준(confidence level): 같은 모집단으로부터 같은 방식으로 얻은, 관심 통계량을 포함할 것으로 예상되는, 신뢰구간의 백분율
- 구간끝점(interval endpoint): 신뢰구간의 최상위, 최하위 끝점

> 예를 들어, 90% 신뢰구간이란 표본통계량의 부트스트랩 표본분포의 90%를 포함하는 구간을 말한다.

- 표본크기 n과 관심 있는 표본통계량이 주어졌을 때, 부트스트랩 신뢰구간을 구하는 법은 다음과 같다.
    - 데이터에서 복원추출 방식으로 크기 n인 표본을 뽑는다. (재표본추출)
    - 재표본추출한 표본에 대해 원하는 통계량을 기록한다.
    - 1~2단계를 R번 반복한다.
    - x% 신뢰구간을 구하기 위해, R개의 재표본 결과로부터 분포의 양쪽 끝에서 [(100-x) / 2]% 만큼 잘라낸다
    - 절단한 점들은 x% 부트스트랩 신뢰구간의 양 끝점이다.

표준정규분포와 QQ plot
- QQ plot은 "표본분포가 정규분포에 얼마나 가까운지를 시각적으로 보여준다."
    - y축은 각 데이터포인트의 z점수(각 데이터포인트에서 평균을 빼고 표준편차로 나눈 것)이고,
    - x축은 정규분포에서의 해당 분위수를 나타낸다.<br><br>
    
표준정규분포에서 기억해야 할 것은..
- "데이터를 표준화한다고 해서 데이터가 정규분포가 되는 것이 아니라, 단지 비교를 목적으로 데이터를 표준정규분포와 같은 척도로 만드는 것이다."

> 정규분포는 불확실성과 변동성에 대한 수학적 근사가 가능하도록 했다는 점에서 중요하다.

> 원시 데이터 자체는 대개 정규분포가 아니지만, 표본들의 평균과 합계, 그리고 오차는 대개 정규분포를 따른다.

> 데이터를 z 점수로 변환하면 데이터를 정규분포와 비교할 수 있게 된다.

#### t 분포
t 분포는 정규분포와 생김새가 비슷하지만, 꼬리 부분이 약간 더 두껍고 길다. (꼬리가 두텁다는 것은 "higher dispersion in variables를 의미한다") 표본통계량의 분포를 설명하는 데 광범위하게 사용된다.
- 표본평균의 분포는 일반적으로 t 분포와 같은 모양이며, 표본 크기에 따라 다른 계열의 t 분포가 있다. ==>> 표본이 클수록 더 정규분포를 닮은 t 분포가 형성된다.
    - t 분포의 가정은: 표본통계량의 분포가 정규분포를 따른다는 것이다.
- 표본평균이 $\bar x$, 표본크기가 $n$, 표본표준편차가 $s$라면, 표본평균 주위의 90% 신뢰구간은 다음과 같다. ((90% 신뢰구간이라는 건.. 재표본추출했을 때 90% 확률로 표본평균(표본통계량)이 이 구간 내에 들어온다는 것이다.))
$$ \bar x \pm t_{n-1}(.05) \times \dfrac{s}{\sqrt n} $$

- 여기서 $t_{n-1}(.05)$ 는 $(n-1)$의 자유도를 갖는 t 분포의 양쪽 끝에서 5%를 잘라내는 t 통계량을 의미한다. => 다시 말해 90% 신뢰구간

> *t 분포는 근본적으로.. 표본분포를 근사화하기 위한 수학적 기법이다. 지금은 컴퓨터를 통해 엄청난 계산을 요하는 재표본추출을 쉽게 할 수 있다.*<br>
> **처음 t 분포를 왜 만들었냐면.. 소표본일 때 정규분포가 잘 들어맞지 않아서 사용한 것이다. 표본 개수가 30개 이상이 되면 정규분포에 가까워지지만, 그 이하에서는 정규분포를 아래로 누른 모양이 된다.**<br><br>

#### t 분포 추가
- 정규분포의 평균을 측정할 때 주로 사용한다.
    - 모평균을 추정/검정하고 싶은데 모표준편차를 모를 때 정규분포 대신 사용할 수 있다.
- 표준정규분포와 유사하게 "0을 중심으로 좌우대칭하는 분포"다.
    - 표준정규분포보다 평평하고 기다란 꼬리를 가진다. (양쪽 꼬리가 두텁다)
    - t분포는 '자유도'라는 모수에 의해 그 모양이 결정된다.(여기서 자유도='표본크기-1')
    - 표본 크기가 적으면 적을수록 분포의 꼬리가 더 평평해진다.

#### 이항분포
이항분포란 각 시행마다 그 성공확률($p$)이 정해져 있을 때, 주어진 시행 횟수($n$) 중에서 성공한 횟수($x$)의 도수분포를 의미한다. $x, n, p$ 값에 따라 다양한 이항분포들이 있다.
- 이항분포의 질문은 예를 들어 이런 것이다: "한 번의 클릭이 판매로 이어질 확률이 0.02일 때, 200회 클릭으로 0회 매출을 관찰할 확률은 얼마인가?"
- 이항분포의 평균은 $n \times p$이고, 분산은 $n \times p(1-p)$이다.
    - 시행횟수가 충분하고 성공확률이 0.5에 가까울 때 이항분포는 사실상 정규분포와 구별이 어렵다.

#### 푸아송(Poisson) 분포

푸아송 분포는 '시간별 혹은 공간별로 사건 발생 정도'를 알고 싶을 때 사용한다. 푸아송 분포는 시간 단위 또는 공간 단위로 표본들을 수집할 때, 그 사건들의 분포를 알려준다.
- 예를 들어.. `rpois(100, lambda = 2)`라는 코드는 lambda=2인 푸아송 분포에서 100개의 난수를 생성한다. e.g. 고객 서비스 센터에 접수되는 문의 전화가 분당 평균 2회이면, 이 코드는 100분을 시뮬레이션하여 100분당 문의 전화 횟수를 알려준다.

> lambda: 단위 시간이나 단위 면적당 사건이 발생하는 비율(평균 사건 수); e.g. 웹사이트에 방문한 방문객 수에 따른 사건 발생 비율, 톨게이트에 들어오는 자동차 수에 따른 사건 발생 비율, 1제곱미터당 건물의 결함 비율, 코드 100줄당 오타 비율 등.

#### 지수(Exponential 분포)

푸아송 분포와 동일한 변수 $\lambda$를 사용하여 '사건과 사건 간의 시간 분포'를 모델링할 수 있다.
- `rexp(n = 100, rate = .2)`라는 함수는 주기별 평균 사건 수가 0.2인 지수분포에서 100개의 난수를 생성한다. e.g. 분당 평균 0.2회의 서비스 문의 전화가 걸려오는 경우, 100분 동안의 서비스 센터 문의 전화 횟수를 시뮬레이션할 수 있다.

> 푸아송과 지수에서 핵심적인 가정은 "$\lambda$가 해당 기간동안 일정하게 유지된다는 가정"이다.


#### 베이불(Weibull) 분포

많은 경우 사건 발생률은 시간에 따라 일정하지 않다. 변화 주기가 일반적인 사건 발생 구간보다 훨씬 길다면, 그냥 비율이 일정한 구간으로 분석을 세분화하면 된다. **그러나** 사건 발생률이 시간에 따라 지속적으로 변한다면 푸아송 분포나 지수 분포는 유효하지 않다.
- 대표적인 예시: 기계 고장! 시간이 지날수록 고장 위험은 증가한다.
- 베이불 분포는 지수분포를 확장한 것으로 $\beta$와 $\eta$라는 두 가지 인수를 갖는다.
    - (1) 형상(shape) 파라미터 $\beta$가 어떻게 지정되느냐에 따라 발생률이 달라진다. $\beta > 1$이면 발생률은 시간에 따라 증가하고, $\beta < 1$이면 감소한다.
    - (2) 푸아송이나 지수 분포처럼 구간당 사건 발생률을 나타내는 $\lambda$가 아니라, '특성 수명'을 표현하는 $\eta$를 사용한다. => 보통 베이불 분포는 사건 발생률 대신 고장 시간 분석에 사용하기 때문에 그렇다.

## 3장 통계적 실험과 유의성 검정

### 3.1 A/B 검정

A/B 검정은 두 처리 방법, 제품, 혹은 절차 중 어느 쪽이 다른 쪽보다 더 우월하다는 것을 입증하기 위해 실험군을 두 그룹으로 나누어 진행하는 실험이다.
- 두 가지 처리법 중 하나는 기준이 되는 기존 방법(대조군)을 쓰고, 나머지 하나는 새로운 방법(처리군)을 쓴다. => 새로운 처리법이 대조군보다 더 낫다는 게 일반적인 가설이 된다.

> 검정통계량(test statistic): 처리 효과를 측정하기 위한 지표

### 3.2 가설검정

가설검정(hypothesis test) 혹은 유의성 검정(significance test)는 "관찰된 효과가 우연에 의한 것인지 여부를 알아내기 위한" 전통적인 통계분석 방법이다.

> 가설검정은 귀무가설이 사실이라고 가정하고 '영모형(null model)', 즉 확률모형(probability model)을 생성하여 관찰한 효과가 해당 모델로부터 합리적으로 나올 수 있는 결과인지 여부를 검증하는 것이다.

### 3.3 재표본추출

재표본추출에는 '부트스트랩(bootstrap)'과 '순열검정(permutation test)'이라는 두 가지 주요 유형이 있다. ((앞에서 다뤘듯이, 부트스트랩은 추정의 신뢰성을 평가하는 데 주로 사용된다. => 표본통계량의 신뢰구간 찾기))
- 순열검정은 두 개 이상의 그룹과 관련된 가설을 검증하는 데 사용한다.
- 순열 절차는 다음과 같다:
    - 여러 그룹의 결과(처리군과 대조군)를 단일 데이터 집합으로 결합한다.
    - 결합한 데이터를 잘 섞은 후, 그룹 A와 동일한 크기의 표본을 무작위로 (비복원) 추출한다.
    - 나머지 데이터에서 그룹 B와 동일한 크기의 샘플을 무작위로 (비복원) 추출한다.
    - C, D 등의 그룹에 대해서도 동일한 작업을 수행한다.
    - 원래 샘플(예를 들면 그룹 비율의 차이)에 대해 구한 통계량 또는 추정치가 무엇이었든 간에 지금 추출한 재표본에 대해 모두 다시 계산하고 기록한다. 이것으로 한 번의 순열 반복이 진행된 것이다.
    - 앞선 단계들을 R번 반복하여 검정통계량의 순열 분포를 얻는다.
> "이제 실험을 통해 관찰했던 그룹 간의 차이점으로 돌아가서 순열 과정에서 얻은 집합에서의 차이와 비교해보자. 관찰된 차이가 순열로 보이는 차이의 집합 안에 잘 들어 있다면, 우리는 어떤 것도 증명할 수 없다. 즉, 관찰된 차이가 우연히 일어날 수 있는 범위 안에 있다는 말이다. 하지만 관찰된 차이가 대부분의 순열 분포 바깥에 있다면, 우리는 이것은 우연 때문이 아니라고 결론 내릴 수 있다. 다시 말해, 통계적으로 유의미하다고 말할 수 있다."

### 3.4 통계적 유의성과 p 값
> - p 값(p-value): 귀무가설을 구체화한 기회 모델이 주어졌을 때, 관측된 결과와 같이 특이하거나 극단적인 결과를 얻을 확률
> - 유의수준/알파(alpha): 실제 결과가 통계적으로 의미 있는 것으로 간주되기 위해, 우연에 의한 기회 결과가 능가해야 하는 '비정상적인' 가능성의 임계 확률
> - 제1종오류(type 1 error): 우연에 의한 효과가 실제 효과라고 잘못 결론 내리는 것 => 귀무가설을 채택해야 하는 상황에서 대립가설을 채택하는 것
> - 제2종오류(type 2 error): 실제 효과를 우연에 의한 효과라고 잘못 결론 내리는 것 => 대립가설을 채택해야 하는 상황에서 귀무가설을 채택하는 것

```python
# 예를 들어 다음과 같은 실험 결과가 있다고 하자
import pandas as pd
result = pd.DataFrame({'가격A': [200, 23539], '가격B': [182, 22406]}, index=['전환', '전환되지 않음'])
result
```

가격 A가 가격 B에 비해 약 5% 정도 우수한 결과(0.8425% 대 0.8057%로 약 0.0368%p 개선)를 보였는데, 이게 통계적으로 유의한 차이일까? 아니면 우연에 의한 차이였을까?
- 여기서 '우연에 의한 차이'란.. **두 전환율 사이에 차이가 없다는 귀무가설의 확률모형을 가지고 생성한 데이터의 랜덤 변이를 의미**한다.
- 다음 순열 절차는 '두 가격이 동일한 전환율을 공유하는지, 이 랜덤 변이가 5%만큼의 차이를 만들어낼 수 있는지'를 묻는 질문에 답을 준다.
    - (1) 모든 표본 정보가 담긴 항아리가 있다고 생각해보자. 그러면 전체 전환율은 45,945개의 0과 382개의 1이므로 0.008246 = 0.8246%라고 할 수 있다.
    - (2) 크기 23,739(가격A)의 표본을 섞어서 뽑고, 그중 1이 몇 개인지 기록한다.
    - (3) 나머지 22,588개(가격B)에서 1의 수를 기록한다.
    - (4) 1의 비율 차이를 기록한다.
    - (5) 2~4단계를 반복한다.
    - (6) 이 차이가 얼마나 자주 >= 0.0368 인가??

위의 경우 순열 분포를 보면 0.0368%의 차이는 랜덤 변이의 범위 내에 있다.<br>
- 순열 검정의 결과로 p값을 추정할 수도 있는데, 이 경우 0.308이 나온다. 그 의미는 "우연히 얻은 결과의 30% 정도가 관찰한 것만큼 극단적이거나 그 이상 극단적인 결과를 얻을 것으로 기대된다"는 말이다.
- 사실 위의 경우에서는 가설이 이항분포를 따르기 때문에 순열검정을 하지 않고 정규분포를 사용하여 p값을 근사할 수도 있다.

> 우리는 p 값의 의미를 "결과가 우연에서 비롯된 확률"로 사용하고 싶어하지만, 그 실제 의미는 **"랜덤 모델이 주어졌을 때 그 결과가 관찰된 결과보다 더 극단적일 확률"** 이다.

### 3.5 t 검정

다양한 유형의 유의성 검정 방법 중 가장 자주 사용되는 것은 t 검정(t-test)이다. (t 검정은 원래 윌리엄 고셋이 단일 표본평균의 분포를 근사화하기 위해 개발한 것이다)<br>
- 1920~30년대 재표본 검정을 위해 수천번 데이터를 섞어서 순열 분포를 만드는 것이 불가능했는데, 통계학자들이 순열 분포에 대한 좋은 근사가 고셋의 t 분포에 기초한 t 검정이라는 것을 발견했다.

> - 검정통계량(test statistic): 관심의 차이 또는 효과에 대한 측정 지표
> - t 통계량(t-statistic): 표준화된 형태의 검정통계량
> - t 분포(t-distribution): 관측된 t 통계량을 비교할 수 있는, 귀무가설에서 파생된 기준 분포<br>

> **모든 유의성 검정은 관심 있는 효과를 측정하기 위한 검정통계량을 지정하고, 관찰된 효과가 정상적인 랜덤 변이의 범위 내에 있는지 여부를 판단하는 데 도움을 준다.**

### 3.6 다중검정

> - 거짓발견비율(False Discovery Rate, FDR): 다중검정에서 1종오류가 발생하는 비율
> - p값 조정(Adjustment of p-value): 동일한 데이터에 대해 다중검정을 수행하는 경우에 필요하다
> - 과대적합(Overfitting): 잡음까지 피팅하는 현상<br><br>


예를 들어 유의수준 0.05에서 20번의 유의성 검정을 수행하면 적어도 하나의 예측변수에서 1종오류를 범할 가능성이 꽤 있다. (0.95^20=0.36, 즉 20번 모두 올바른 검정 결과를 보일 확률이고.. 1 - 0.36 = 0.64는 적어도 하나의 예측값에서 1종오류를 범할 확률이다.)

## 3.8 분산분석

여러 그룹의 수치 데이터를 서로 비교할 때, 여러 그룹 간의 통계적으로 유의미한 차이를 검정하는 통계적 절차를 분산분석(ANalysis Of VAriance, ANOVA)이라고 부른다.

> - 쌍별 비교(pairwise comparison): 여러 그룹 중 두 그룹 간의 (예를 들면 평균에 대한) 가설검정
> - 총괄검정(omnibus test): 여러 그룹 평균들의 전체 분산에 관한 단일 가설검정
> - 분산분해(decomposition of variance): 구성 요소 분리. 예를 들면 전체 평균, 처리 평균, 잔차 오차로부터 개별 값들에 대한 기여를 뜻한다.
> - F 통계량: 그룹 평균 간의 차이가 랜덤 모델에서 예상되는 것보다 벗어나는 정도를 측정하는 표준화된 통계량
> - SS(Sum of Squares): 어떤 평균으로부터의 편차들의 제곱합<br><br>


- ANOVA의 토대가 되는 재표본추출 과정은 다음과 같다: (e.g. 5개 값을 갖는 그룹 네 개를 비교할 때)
    1. 모든 데이터를 한 상자에 모은다.
    2. 5개의 값을 갖는 4개의 재표본을 섞어서 추출한다.
    3. 각 그룹의 평균을 기록한다.
    4. 네 그룹 평균 사이의 분산을 기록한다.
    5. 2~4단계를 여러 번 반복한다.
    6. 재표집된 분산이 관찰된 변화를 초과한 시간은 어느 정도일까? 이것이 바로 p값이다.
